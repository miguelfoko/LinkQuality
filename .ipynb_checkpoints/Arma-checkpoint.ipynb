```python
import pandas as pd
%matplotlib inline

import matplotlib.pyplot as plt
pd.options.display.max_rows=10
from datetime import datetime

import seaborn
import xgboost
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import scale
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import preprocessing
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification

import scipy
from scipy.stats import spearmanr
from pylab import rcParams
import seaborn as sb
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.stattools import adfuller, arma_order_select_ic
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

#rcParams['figure.figsize']=5,4
#sb.set_style('whitegrid')

import numpy as np
df = pd.read_table("grenoble_7.k7",sep = ',',header = 0)
class_names=['Bad','Intermediate','Good']


def plot_ar1():
    n_samples = int(1000)
    a = 0.6
    x = w = np.random.normal(size=n_samples)
    for t in range(n_samples):
        x[t] = a*x[t-1] + w[t]

    _=seaborn.tsplot(x)

    
def compute_alpha_ar1():
    n_samples = int(1000)
    a = 0.6
    x = w = np.random.normal(size=n_samples)
    for t in range(n_samples):
        x[t] = a*x[t-1] + w[t]
        
    mdl = sm.tsa.AR(x).fit(maxlag=30, ic='aic', trend='nc')
    est_order = sm.tsa.AR(x).select_order( maxlag=30, ic='aic', trend='nc')

    true_order = 1
    print('\nalpha estimate: ',mdl.params,' | order_estimate: ',est_order)
    print('\ntrue alpha = %s | true order = %s'%(a, true_order))

def plot_ar2(df):
    training_set=df.iloc[:,[0,5]]
    training_set=pd.DataFrame(training_set)
    training_set.index=pd.to_datetime(df.iloc[:,0])
    print("Les types: ===>",training_set.dtypes)
    #training_set.reset_index().set_index('datetime').resample('H').mean()
    #training_set=pd.DataFrame(training_set)
    #training_set.plot()
    results = adfuller(training_set.iloc[:,1].values, autolag='AIC')
    print("Result[1]===> ",results[1])
    if results[1]<=0.05:
        print("Reject Null hypothesis, the series are stationary")
    else:
        print("Do not reject Null, the series are not stationary")
    
    model=ARMA(training_set.pdr,order=(3,0))
    res=model.fit()
    res= res.predict()

    print("Heyyyyyyyyyyyyy ===>",res)
    
    plt.ylabel('Baysian Information Criterion')
    plt.plot(res)
    plt.show()
    print("According to Baysian Information criteria, we can use (ARMA(3,0) model)")
    print ("min: ",training_set.index.min())
    print ("max: ",training_set.index.max())
    
def plot_ar3(df):
    df=df.set_index('datetime')
    ts=df['pdr']
    
    #Plot the time serie
    #plt.plot(ts)
    
    #Check the stationarity of the time serie
    #1 Determining rolling statistics

    #rolmean=ts.rolling(500).mean()
    rolmean=ts.rolling(window=500).mean()
    rolstd=ts.rolling(500).std()
    
    #2Plot rolling statistics
    orig=plt.plot(ts,color='blue',label='Original')
    mean=plt.plot(rolmean,color='red',label='Rolling Mean')
    std=plt.plot(rolstd,color='black',label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standart Deviation')
    plt.show(block=False)
    #3Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller test:')
    dftest=adfuller(ts,autolag='AIC')
    dfoutput=pd.Series(dftest[0:4],index=['Test Statistic','p-value','Lages used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key]=value
        print (dfoutput)

        
def studies(df,src,dst,ch):
    #df=df.set_index('datetime')
    df=df.loc[((df['src']==src)&(df['dst']==dst))]
    taille=list()
    for i in range(len(df)):
        k=i+1
        taille.append(k)
    df['taille']=taille
    df=df.set_index('taille')
    
    ts=df['pdr']
    
    #rolmean=ts.rolling(500).mean()
    rolmean=ts.rolling(window=50).mean()
    rolstd=ts.rolling(50).std()
    
    #2Plot rolling statistics
    #orig=plt.plot(ts,color='blue',label='Original')
    print(ts)
    mean=plt.plot(rolmean,color='red',label='Rolling Mean')
    #std=plt.plot(rolstd,color='black',label='Rolling Std')
    plt.legend(loc='best')
    title1='Time Series PDR '+str(src)+"==>"+str(dst)
    plt.title(title1)
    #plt.show(block=False)
    title="SOCALE\\TS\\PDR\\TS_LINK_"+str(src)+"==="+str(dst)+".png"
    plt.savefig(title, format='png', bbox_inches='tight', pad_inches=0)
    plt.clf()
    
    
    
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return np.array(diff)


def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]

def puissant_arma(df,src,dst):
    df=df.loc[((df['src']==src)&(df['dst']==dst))]
    df=df.set_index('datetime')
    df.index=pd.to_datetime(df.index)
    print(df)
    #ts=df.loc[(df['channel']==11)]
    ts=df['pdr']
    X=ts.values
    minutes_in_hour=1
    dif=difference(X,minutes_in_hour)
    #start_index=len(dif)
    start_index=0
    end_index=len(dif)
    q=1
    p=1
    model=ARMA(ts,order=(p,q))
    model_fit=model.fit(disp=0)
    forecast=model_fit.predict(start=start_index,end=end_index)
    score=model_fit.score(start=start_index,end=end_index)
    
    #model=ARMA(ts,order=(4,3))
    #model_fit=model.fit(disp=0)
    #print(model_fit.summary())
    from statsmodels.tsa.seasonal import seasonal_decompose
    result = seasonal_decompose(ts, model='additive', freq=16)
    result.plot()
    title1='Original Seasonality PDR '+str(src)+"==>"+str(dst)
    #title1='Seasonality  RSSI '+str(src)+"==>"+str(dst)
    plt.title(title1)
    title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\CHOOSE\\SEASONALITY_LINK_ORIGINAL"+str(src)+"==="+str(dst)+".png"
    plt.savefig(title, format='png', bbox_inches='tight', pad_inches=0)
    plt.clf()
    
    result1 = seasonal_decompose(forecast, model='additive', freq=16)
    result1.plot()
    
    title1='ARMA Seasonality PDR '+str(src)+"==>"+str(dst)
    #title1='Seasonality  RSSI '+str(src)+"==>"+str(dst)
    plt.title(title1)

    title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\CHOOSE\\SEASONALITY_LINK_ARMA("+str(p)+","+str(q)+")_"+str(src)+"==="+str(dst)+".png"
    #title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\SEASONALITY\\PDR\\SEASONALITY_LINK_"+str(src)+"==="+str(dst)+".png"
    #title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\SEASONALITY\\RSSI\\PACF_LINK_"+str(src)+"==="+str(dst)+".png"
    plt.savefig(title, format='png', bbox_inches='tight', pad_inches=0)
    plt.clf()

def my_plot_predict(df,src,dst):
    df=df.loc[((df['src']==src)&(df['dst']==dst))]
    df=df.set_index('datetime')
    df.index=pd.to_datetime(df.index)
    ts=df['pdr']
    X=ts.values
    minutes_in_hour=1
    dif=difference(X,minutes_in_hour)
    p=1
    q=0
    start_index=1
    end_index=len(dif)
    res=sm.tsa.ARMA(ts,(p,q)).fit()
    
    #Afficher les scores ici
    
    fig,ax=plt.subplots()
    ax=ts.plot(ax=ax,color='red',label='Original PDR')
    fig=res.plot_predict(start_index,end_index,dynamic=True,ax=ax,plot_insample=False)
    
    title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\CHOOSE\\PDR\\FORECASTED_LINK_ARMA("+str(p)+","+str(q)+")_"+str(src)+"==="+str(dst)+".png"
    #title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\CHOOSE\\RSSI\\FORECASTED_LINK_ARMA("+str(p)+","+str(q)+")_"+str(src)+"==="+str(dst)+".png"
    #title="C:\\Users\WSN-LINK\\Documents\\TEST\\SOCALE\\SEASONALITY\\RSSI\\PACF_LINK_"+str(src)+"==="+str(dst)+".png"
    plt.savefig(title, format='png', bbox_inches='tight', pad_inches=0)
    plt.clf()

    
my_plot_predict(df,4,32)
#puissant_arma(df,4,32)   
#plot_ar3(df)
#studies(df,0,12,16)
```
